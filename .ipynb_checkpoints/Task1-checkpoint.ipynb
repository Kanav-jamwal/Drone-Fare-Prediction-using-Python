#### 9.2 Data Cleansing of Regions

##### 9.2.1 Orgin Region 

Approach: 

    1.Find the mean of the Lat-Lon of the all regions.
    2.Find the distance between all the lat-lon with median for each region-Determining the farthest value(cosidering-3 standard devaitions)
    3.Create a list of all such Lat-Lon
    4.Determine the distance between lat-lon (created in step 4) with mean Lat-Lon
    5.Find min Distance between them and assign region corresponding to mean
    6. Continue the same for all the regions.

```python
# Function to Calculate the Lat-lon Distance
def lat_log_dist(lat1,lon1,lat2,lon2):
    R = 6378
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
```

```python
#Function to Calculate Closest lat-lon 
def closest(data, v):
    z=min(data, key=lambda p: lat_log_dist(v['Origin Latitude'],v['Origin Longitude'],p['lat'],p['lon']))
    return data.index(z)
```

```python
#Calculating Mean for each Region 
d={} 
mean_list=[]
for i in range(1,45):
    m_lat=drone_file_df['Origin Latitude'][drone_file_df['Origin Region']==i].mean()
    m_lon=drone_file_df['Origin Longitude'][drone_file_df['Origin Region']==i].mean()
    d['lat']=m_lat
    d['lon']=m_lon
    mean_list.append(d.copy())
```

```python
#calculating lat-lon which are doesn't belong to group
ouliers_list=[]
for i in range(1,45):
    df_filtered=drone_file_df[drone_file_df['Origin Region']==i] 
    z=[]
    m_lat=drone_file_df['Origin Latitude'][drone_file_df['Origin Region']==i].mean()
    m_lon=drone_file_df['Origin Longitude'][drone_file_df['Origin Region']==i].mean()
    for index, row in df_filtered.iterrows():    
        O_lat = radians(drone_file_df['Origin Latitude'][index])
        O_lon = radians(drone_file_df['Origin Longitude'][index])
        z.append(lat_log_dist(O_lat,O_lon,m_lat,m_lon))

    df_filtered['dist']=z
       
    o=df_filtered[~(np.abs(df_filtered.dist-df_filtered.dist.mean())<=(3*df_filtered.dist.std()))]
    ouliers_list.append(o)
```

```python
# Finding the Distance between outliers and Mean 
ouliers_df = pd.concat(ouliers_list)   
ouliers_df=ouliers_df.reset_index().drop(['index'],axis=1)
for x,y in ouliers_df.iterrows():
    ouliers_df['Origin Region'][x]=closest(mean_list,y)+1
```

```python
#Adding the corrected value to original Dataframe  
n=list(ouliers_df.Id)
ouliers_df=ouliers_df.drop(['dist'],axis=1)
drone_file_df = drone_file_df[~drone_file_df['Id'].isin(n)]
drone_file_df=pd.concat([drone_file_df,ouliers_df])
drone_file_df=drone_file_df.reset_index().drop(['index'],axis=1)
```

Determinig the records which have null values for Origin region an imputing the correct value based on the minimum Distance

```python
Null_origin_region=drone_file_df[drone_file_df['Origin Region'].isnull()].reset_index().drop(['index'],axis=1)
for x,y in Null_origin_region.iterrows():
    Null_origin_region['Origin Region'][x]=closest(mean_list,y)+1
n=list(Null_origin_region.Id)
#Null_origin_region=Null_origin_region.drop(['dist'],axis=1)
drone_file_df = drone_file_df[~drone_file_df['Id'].isin(n)]
drone_file_df=pd.concat([drone_file_df,ouliers_df])
drone_file_df=drone_file_df.reset_index().drop(['index'],axis=1)
```

##### 9.2.2 Destination Region 

```python
def closest_dest(data, v):
    z=min(data, key=lambda p: lat_log_dist(v['Destination Latitude'],v['Destination Longitude'],p['lat'],p['lon']))
    return data.index(z)
```

```python
#Identifying Outliers list
dd={} 
mean_list_d=[]
for i in range(1,45):
    m_lat_d=drone_file_df['Destination Latitude'][drone_file_df['Destination Region']==i].mean()
    m_lon_d=drone_file_df['Destination Longitude'][drone_file_df['Destination Region']==i].mean()
    dd['lat']=m_lat_d
    dd['lon']=m_lon_d
    mean_list_d.append(dd.copy())
ouliers_list_d=[]
for i in range(1,45):
    df_filtered_d=drone_file_df[drone_file_df['Destination Region']==i] 
    y=[]
    m_lat_d=drone_file_df['Destination Latitude'][drone_file_df['Destination Region']==i].mean()
    m_lon_d=drone_file_df['Destination Longitude'][drone_file_df['Destination Region']==i].mean()
    for index, row in df_filtered_d.iterrows():    
        O_lat_d = radians(drone_file_df['Destination Latitude'][index])
        O_lon_d = radians(drone_file_df['Destination Longitude'][index])
        y.append(lat_log_dist(O_lat_d,O_lon_d,m_lat_d,m_lon_d))
    
    df_filtered_d['dist']=y
       
    o=df_filtered_d[~(np.abs(df_filtered_d.dist-df_filtered_d.dist.mean())<=(3*df_filtered_d.dist.std()))]
    ouliers_list_d.append(o)
```

```python
#Finding the Closest
ouliers_df_d = pd.concat(ouliers_list_d)   
ouliers_df_d=ouliers_df_d.reset_index().drop(['index'],axis=1)
for x,y in ouliers_df_d.iterrows():
    ouliers_df_d['Destination Region'][x]=closest_dest(mean_list_d,y)+1
```

```python
#Adding back to dataframe
n=list(ouliers_df_d.Id)
ouliers_df_d=ouliers_df_d.drop(['dist'],axis=1)
drone_file_df = drone_file_df[~drone_file_df['Id'].isin(n)]
drone_file_df=pd.concat([drone_file_df,ouliers_df_d])
drone_file_df=drone_file_df.reset_index().drop(['index'],axis=1)
```

```python
Null_dest_region=drone_file_df[drone_file_df['Destination Region'].isnull()].reset_index().drop(['index'],axis=1)
for x,y in Null_dest_region.iterrows():
    Null_dest_region['Destination Region'][x]=closest_dest(mean_list_d,y)+1
n=list(Null_dest_region.Id)
drone_file_df = drone_file_df[~drone_file_df['Id'].isin(n)]
drone_file_df=pd.concat([drone_file_df,ouliers_df_d])
drone_file_df=drone_file_df.reset_index().drop(['index'],axis=1)
```

#### 10.2  Calculating Difference between Journey Distance and Calculated Latitude longitude Distance

We know Journey Distance must be distance between latitudes and longitudes. Hence we calculate the difference between latitudes and longitudes of origin and distance

```python
def lat_log_dist(lat1,lon1,lat2,lon2):
    R = 6378
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
    
values=[]
for index, row in drone_file_df.iterrows():    
    O_lat = radians(drone_file_df['Origin Latitude'][index])
    O_lon = radians(drone_file_df['Origin Longitude'][index])
    D_lat = radians(drone_file_df['Destination Latitude'][index])
    D_lon = radians(drone_file_df['Destination Longitude'][index])
    values.append(round(lat_log_dist(O_lat,O_lon,D_lat,D_lon),3))

drone_file_df['lat_log_dist']=values
```

Now we shall compare it with actual Journey Distance 

```python
drone_file_df.plot(kind='scatter',x='Journey Distance',y='lat_log_dist',figsize=(15,10))
```

This Graphs Shows there are minute differeces in Journey Distance and Latitude-Longitude Distance calculated. We can also prove this by using non-graphical method

```python
x=[]
for index,row in drone_file_df.iterrows():
    if drone_file_df['lat_log_dist'][index] != round(drone_file_df['Journey Distance'][index],3):
        x.append('0')
    else:
        x.append('1')

drone_file_df['lat_log_compare']=x
```

```python
drone_file_df[drone_file_df['lat_log_compare']=='0'].count()
```

From this we can infer that there are few journey Distances, which are not in sync with distance between latitudes and longitudes. Specifically 15 of them.

### 13. Calculating the Speed of each Drone.

Let x,y,z be the speeds of drones while take off, Landing and enroute. We know that it takes 5 min to take off and 5 min to land Hence Distance(D) travelled during that time will be 5x and 5y.Let 't' be the time taken to travel from origin to Destination.

Then D =(5x+5y+(t-10)z)/60
 
Let 5x+5y by 'B'. Then D=(y+(t-10)z)/60 then 

We solve the equations by taking two samples where D=>Davag and t will be tavg. By Solving these two equations we find the value of Z. We do this for each drone type only if the difference between delivery time departure time as correct. Knowing the value of Z, we substitue it back one the eqautions to find the value of 'B'. 

After finding the value of B,
we substitue this back into the equation below to find the value of t. 
t=10+(D-B)/Z 

Each Drone type will have its own Speeds. Based on the Speed we determine  the typ of drones. 

And from the type of drone we calculate its travel time. And finally calculate delivery time based on departure time + travel time.

```python
#Number of dates which has to reversed
count=0
for i,j in drone_file_df.iterrows():
    if not math.isnan(drone_file_df['Travel Time'][i]):   
        d=j['Departure Time'].split(":")
        a=j['Delivery Time'].split(":")
        
        t=int(round(j['Travel Time']/60,0))
        if not int(d[0]) +int(t) >=24 and t!=0:
            diff=int(a[0])-int(d[0])
            if diff==-1*int(t):
                count+=1
```

```python
print(count)
print("So there are 21 cases where we have delivery time less that departure time. Hence we consider inverting them")
```

```python
#Inverting the delivery time and departure time
for i,j in drone_file_df.iterrows():
    if not math.isnan(drone_file_df['Travel Time'][i]):   
        d=j['Departure Time'].split(":")
        a=j['Delivery Time'].split(":")
        
        t=int(round(j['Travel Time']/60,0))
        if not int(d[0]) +int(t) >=24 and t!=0:
            diff=int(a[0])-int(d[0])
            if diff==-1*int(t):
                drone_file_df['Departure Time'][i],drone_file_df['Delivery Time'][i]=drone_file_df['Delivery Time'][i],drone_file_df['Departure Time'][i]
```

Now we shall calculate the difference between two departure time and delivery time which has be equal to given travel time

```python
#Calculating Differece
values=[]
for index, row in drone_file_df.iterrows():
    FMT = '%H:%M:%S'
    a=datetime.strptime(drone_file_df['Delivery Time'][index],FMT) - datetime.strptime(drone_file_df['Departure Time'][index],FMT)
    if a.days < 0:
        a = timedelta(days=0,seconds=a.seconds)
        values.append(round(a.total_seconds()/60,2))
    else:
        values.append(round(a.total_seconds()/60,2))
drone_file_df['time_diff']=values # adding it to Dataframe as a new column 
```
